## Data wrangling

### Subtask:
Data Wrangling and Feature Engineering


**Reasoning**:
Clean the currency formatting from the numerical columns in `df_with_year`, convert them to numeric types, and then engineer new features representing year-over-year changes and financial ratios. Handle any missing values introduced during these calculations.



```python
import pandas as pd
import numpy as np

# Numerical columns to clean and convert
numerical_cols = ['persediaan_awal', 'persediaan_akhir', 'hpp_rawat_inap', 'hpp_rawat_jalan', 'total_hpp',
                  'pendapatan_operasional', 'pendapatan_penjualan', 'ppn_keluaran', 'ppn_masukan', 'ppn_netto',
                  'laba_kotor_penjualan', 'beban_operasional', 'laba_kotor_operasional', 'laba_sebelum_pajak',
                  'laba_kena_pajak', 'laba_bersih_normal', 'laba_bersih_tax_holiday']

# Clean currency formatting and convert to numeric
for col in numerical_cols:
    df_with_year[col] = pd.to_numeric(df_with_year[col].astype(str).str.replace('.', '', regex=False).str.replace(',', '.', regex=False), errors='coerce')

# Calculate year-over-year percentage change
for col in ['pendapatan_operasional', 'laba_kotor_operasional', 'laba_bersih_normal']:
    df_with_year[f'{col}_yoy_change'] = df_with_year[col].pct_change() * 100

# Calculate financial ratios
df_with_year['gross_profit_margin'] = (df_with_year['laba_kotor_operasional'] / df_with_year['pendapatan_operasional']) * 100
df_with_year['operating_profit_margin'] = (df_with_year['laba_sebelum_pajak'] / df_with_year['pendapatan_operasional']) * 100
df_with_year['net_profit_margin'] = (df_with_year['laba_bersih_normal'] / df_with_year['pendapatan_operasional']) * 100

# Handle missing values after feature engineering
# Using forward fill for year-over-year changes and ratios, as they are time-series related
for col in [f'{col}_yoy_change' for col in ['pendapatan_operasional', 'laba_kotor_operasional', 'laba_bersih_normal']] + ['gross_profit_margin', 'operating_profit_margin', 'net_profit_margin']:
    df_with_year[col] = df_with_year[col].fillna(method='ffill')

display(df_with_year.head())
display(df_with_year.tail())
```
